```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
solvay = cv2.imread('solvay_conference.jpg')
nadia = cv2.imread('Nadia_Murad.jpg')
```

```python
plt.imshow(solvay, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
  face_img = img.copy()
  face_rects = face_cascade.detectMultiScale(face_img)
  for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
  return face_img
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
def detect_face(img):
  face_img = img.copy()
  face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
  for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
  return face_img
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eye(img):
  eye_img = img.copy()
  eye_rects = eye_cascade.detectMultiScale(eye_img,scaleFactor=1.2,minNeighbors=5)
  for (x,y,w,h) in eye_rects:
    cv2.rectangle(eye_img,(x,y),(x+w,y+h),(255,255,255),10)
  return eye_img
```

```python
result = detect_eye(nadia)
plt.imshow(result,cmap='gray')
```
